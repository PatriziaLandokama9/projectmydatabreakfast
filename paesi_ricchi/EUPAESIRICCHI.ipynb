```python
#statistica descrittiva 
import pandas as pd
#crea un dataFrame 
ricchi=pd.DataFrame({
    "Paese":["Lussemburgo","Irlanda","Svizzera","Norvegia","Islanda","Danimarca","Paesi Bassi","San Marino","Austria","Svezia","Belgio","Finlandia","Germania","Regno Unito"],
    "Salario":[131.380, 106.060, 105.670,94.660, 84.590, 68.900, 63.750, 59.410, 59.230, 58.530, 55.540, 55.130, 54.290, 51.070,]
})

media = ricchi['Salario'].mean()
mediana= ricchi['Salario'].median()
devazione_standard= ricchi['Salario'].std(ddof=0)
primo_quartile=ricchi['Salario'].quantile(0.25)
terzo_quartile=ricchi['Salario'].quantile(0.75)
massimo=ricchi['Salario'].max()
minimo=ricchi['Salario'].min()
massimo_abs = abs(ricchi['Salario']).max()

print(f"calcolare la media:{media}")
print(f"calcolare la mediana:{mediana}")
print(f"calcolare la devazione_standard:{devazione_standard}")
print(f"calcolare la primo_quartile:{primo_quartile}")
print(f"calcolare la terzo_quartile:{terzo_quartile}")
print(f"calcolare la massimo:{massimo}")
print(f"calcolare la minimo:{minimo}")
print(f"calcolare la massimo_abs:{massimo_abs}")
```

```python
# Scikit-learn preprocessors
x= ricchi[['Salario']].values
Standard = StandardScaler()
Robust = RobustScaler()
MinMax = MinMaxScaler()
MaxAbs = MaxAbsScaler()

x1 = Standard.fit_transform(x)
x2 = Robust.fit_transform(x)
x3 = MinMax.fit_transform(x)
x4 = MaxAbs.fit_transform(x)

# Calcolo indici di Asimmetria e Curtosi
ricchi['Salario_standardScaler'] = Standard.fit_transform(ricchi[['Salario']])
asimmetriaFisher = ricchi['Salario_standardScaler'].pow(3).mean()
Curtosi = ricchi['Salario_standardScaler'].pow(4).mean()

print("Asimmetria di Fisher:", asimmetriaFisher)
print("Curtosi:", Curtosi)
```

```python
# Calcolo con scipy
asimmetria_scipy = scipy.stats.skew(ricchi['Salario'])
curtosi_scipy = scipy.stats.kurtosis(ricchi['Salario'])

print("Asimmetria (scipy):", asimmetria_scipy)
print("Curtosi (scipy):", curtosi_scipy)
```

```python
# Visualizzazione dei risultati dei preprocessors
print("StandardScaler:", x1)
print("RobustScaler:", x2)
print("MinMaxScaler:", x3)
print("MaxAbsScaler:", x4)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dati di esempio
data = [131.380, 106.060, 105.670, 94.660, 84.590, 68.900, 63.750, 59.410, 59.230, 58.530, 55.540, 55.130, 54.290, 51.070]

# Calcoli statistici
media = 74.87214285714285
mediana = 61.58

# Creazione del boxplot verticale
plt.figure(figsize=(12, 6))

# Grafico a boxplot verticale
plt.subplot(1, 2, 1) # 1 SAREBBE RIGA 2 SAREBBE COLLONA 1SAREBBE PRIMO GRAFICO
plt.boxplot(data, vert=True, patch_artist=True, boxprops=dict(facecolor='orange'))
plt.axhline(media, color='r', linestyle='--', label='Media: {:.2f}'.format(media))
plt.axhline(mediana, color='g', linestyle='-', label='Mediana: {:.2f}'.format(mediana))
plt.title("Boxplot Verticale con Media e Mediana")
plt.xlabel("Gruppo")
plt.ylabel("Valori")
plt.legend()

# Grafico a boxplot orizzontale
plt.subplot(1, 2, 2) # 1 SAREBBE RIGA 2 SAREBBE COLLONA 2 SAREBBE SECONDO GRAFICO
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='orange'))
plt.axvline(media, color='r', linestyle='--', label='Media: {:.2f}'.format(media))
plt.axvline(mediana, color='g', linestyle='-', label='Mediana: {:.2f}'.format(mediana))
plt.title("Boxplot Orizzontale con Media e Mediana")
plt.xlabel("Valori")
plt.ylabel("Gruppo")
plt.legend()

# Mostrare i grafici
plt.tight_layout()
plt.show()
```
