

📊 Creazione di un DataFrame con Pandas
Questo semplice progetto mostra come creare un DataFrame utilizzando la libreria Pandas in Python, a partire da un dizionario di dati. Ogni chiave rappresenta una colonna e ogni lista di valori rappresenta le righe del dataset.

✅ Obiettivo
Illustrare la base di partenza per l'inserimento di dati in Pandas, utile per chi sta iniziando con l'analisi dei dati o la manipolazione di dataset tabellari.

📌 Descrizione del contenuto
Il dataset contiene:

parolachiave1: Nomi di persone

parolachiave2: Età

parolachiave3: Città di residenza

parolachiave4: Stipendio mensile

parolachiave5: Data di assunzione (in formato YYYY-MM-DD)

parolachiave6id: ID univoci generati a partire dallo stipendio

```python
## come scrivere pandas su pyhton 
import pandas as pd

# Creazione di un DataFrame da un dizionario

data = {
    "parolachiave1": ["Anna", "Luca", "Marco", "Giulia"],  # Nomi
    "parolachiave2": [25, 30, 28, 22],                     # Età
    "parolachiave3": ["Roma", "Milano", "Torino", "Napoli"],  # Città
    "parolachiave4": [2500, 3000, 2700, 2200],             # Stipendio
    "parolachiave5": ["2025-05-02", "2025-01-30", "2025-10-27", "2025-10-22"],  # Date corrette
    "parolachiave6id": ["id_2500", "id_3000", "id_2700", "id_2200"]  # ID come stringhe

}




# Creazione del DataFrame
df = pd.DataFrame(data)

# Visualizzazione del DataFrame
print(df)
```
